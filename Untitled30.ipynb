```python
import networkx as nx
import matplotlib.pyplot as plt
import math
import pandas as pd

def create_graph(n, k):
    G = nx.Graph()


    for i in range(1, n + 1):
        G.add_node(i)

    for i in range(1, n + 1):
        for j in range(i + 1, n + 1):
            if i % k != j % k:
                G.add_edge(i, j)

    nx.draw(G, with_labels=True, node_color="orange", node_size=300,
            font_color="white", font_size=10, font_weight="bold", width=0.5)
    plt.margins(0.2)
    plt.show()

    print(f"For n={n}, k={k}:")
    print(f"The number of edges are {len(G.edges)}")
    print(f"Expected edges from the formula: {int(((k-1)/k) * ((n * (n-1)) / 2))}")

    return G

def compute_distribution(n, k):
    ll = k - (n % k)
    lu = n % k
    l = [0] * (k - 2)

    for i in range(3, k + 1):
        for a in range(0, min(i, ll) + 1):
            j = i - a
            if 0 <= j <= min(i, lu) and j + a == i:
                l[i - 3] += (math.factorial(ll) / (math.factorial(a) * math.factorial(ll - a))) * pow(n // k, a) * \
                            (math.factorial(lu) / (math.factorial(j) * math.factorial(lu - j))) * pow(n // k + 1, j)

    f=[]
    for i in range(3,k+1):
      f.append(i)
    plt.bar(f, l, width=0.05)
    plt.xlabel("Edge Group Size")
    plt.ylabel("Frequency")
    plt.title(f"Distribution for n={n}, k={k}")
    plt.show()

points = [(40, 7), (20, 5), (10, 3)]

for n, k in points:
    create_graph(n, k)
    compute_distribution(n, k)
```
