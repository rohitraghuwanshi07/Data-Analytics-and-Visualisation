```python
import networkx as nx
import matplotlib.pyplot as plt
import random as rand
import numpy as np
```

```python
G = nx.Graph()
n = 40
k = 7

for i in range(1,n+1):
  G.add_node(i)


nx.draw(G, with_labels = True)
plt.show()
```

```python
rem = [0]*7
for i in range(1, n+1):
  rem[i%k] += 1
  for j in range(1, n+1):
    if i%k != j%k:
      G.add_edge(i,j)

nx.draw(G, with_labels = True)
plt.show()
print(rem)
print(G.number_of_edges())
```
